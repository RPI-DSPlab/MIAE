# Visualizing the data

```python
import os.path

from mia.eval_methods import sample_metric_eval as eval
from mia.utils.datasets.loader import load_dataset

if not os.path.exists("datasets"):
    os.mkdir("datasets")

dataset = load_dataset("cifar10",
                       "datasets",
                       target_transform=None,
                       train_transform=None,
                       test_transform=None)
cifar10_train = dataset.train_set
cifar10_test = dataset.test_set
```

## plot 240 epoch's results

```python
pd_train_vgg16_240epoch = eval.read_metric("metrics-data/pd_il_cs_vgg16_240epoch/pd_results/train_vgg16.json")
pd_test_vgg16_240epoch = eval.read_metric('metrics-data/pd_il_cs_vgg16_240epoch/pd_results/test_vgg16.json')
il_train_vgg16_240epoch = eval.read_metric("metrics-data/pd_il_cs_vgg16_240epoch/il_results/train_vgg16.json")
il_test_vgg16_240epoch = eval.read_metric('metrics-data/pd_il_cs_vgg16_240epoch/il_results/test_vgg16.json')

pd_vgg16_240epoch = eval.SampleMetric(pd_train_vgg16_240epoch, pd_test_vgg16_240epoch, cifar10_train, cifar10_test)
il_vgg16_240epoch = eval.SampleMetric(il_train_vgg16_240epoch, il_test_vgg16_240epoch, cifar10_train, cifar10_test)

pd_vgg16_240epoch.print_score_len()
il_vgg16_240epoch.print_score_len()

pd_vgg16_240distri = pd_vgg16_240epoch.plot_score_distribution()
pd_vgg16_240distri.title("VGG16 240 epoch's score distribution (PD)")
pd_vgg16_240distri.show()

il_vgg16_240distri = il_vgg16_240epoch.plot_score_distribution()
il_vgg16_240distri.title("VGG16 240 epoch's score distribution (IL)")
il_vgg16_240distri.show()
```
